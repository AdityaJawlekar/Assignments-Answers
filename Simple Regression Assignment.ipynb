```python
import pandas as pd
import numpy as np
import pandas_profiling as pp
import sweetviz as sv
import seaborn as sns
```

# Q1 Delivery Time

```python
data = pd.read_csv('delivery_time.csv')
data
```

# # EDA

### Outlier Detection

```python
data.boxplot(column=['Sorting Time'])
```

```python
## No outlier present..
```

### Scatter Plot and Correlation Analysis

```python
## To plot default pairplot and distplot
sns.pairplot(data)
```

```python
sns.distplot(data['Delivery Time'])
```

```python
sns.distplot(data['Sorting Time'])
```

```python
## From Scatter Plot and dist plot it is clear that both 'Delivery Time' and 'Sorting Time' are linerly distributed in positive direction
```

```python
## To find correlation
data.corr()
```

```python
## Correlation between delivery and sorting time is 0.82 so it is strongly correlated..
```

### Fitting a Linear Regression Model

```python
sns.regplot(x="Sorting Time", y="Delivery Time", data=data)
```

```python
import statsmodels.api as smf
X = data['Sorting Time']
Y = data['Delivery Time']
model = smf.OLS(Y, X).fit()
```

```python
print(model.summary())
```

### Predicting new values

```python
#Predict for 15 and 48 sorting time
newdata=pd.Series([15, 48])
data_pred=pd.DataFrame(newdata,columns=['Sorting Time'])
model.predict(data_pred)
```

# Q2 Salary Hike

```python
salary_data = pd.read_csv('Salary_Data.csv')
salary_data
```

# # EDA

### Outlier Detection

```python
salary_data.boxplot(column=['YearsExperience'])
```

```python
## No outlier present..
```

### Scatter Plot and Correlation Analysis

```python
## To plot default pairplot and distplot
sns.pairplot(salary_data)
```

```python
sns.distplot(salary_data['YearsExperience'])
```

```python
sns.distplot(salary_data['Salary'])
```

```python
## From Scatter Plot and dist plot it is clear that both 'Delivery Time' and 'Sorting Time' are linerly distributed in positive direction
```

```python
## To find correlation
salary_data.corr()
```

```python
## Correlation between salary and years of experience time is 0.978 so it is strongly correlated..
```

### Fitting a Linear Regression Model

```python
sns.regplot(x="YearsExperience", y="Salary", data=salary_data)
```

```python
import statsmodels.api as smf
X1 = salary_data['YearsExperience']
Y1 = salary_data['Salary']
salary_model = smf.OLS(Y1, X1).fit()
```

```python
print(salary_model.summary())
```

### Predicting new values

```python
#Predict for 1o and 25 years of experience
salary_newdata=pd.Series([10, 25])
salary_data_pred=pd.DataFrame(salary_newdata,columns=['YearsExperience'])
salary_model.predict(salary_data_pred)
```
