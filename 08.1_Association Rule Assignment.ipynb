# Q1. Prepare rules for the 'Book' data set.

## Q1.1) Try different values of support and confidence. Observe the change in number of rules for different support,confidence values.

## Imoprting Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori,association_rules
```

## Loading Data Set

```python
book_data = pd.read_csv('book.csv')
book_data.head()
```

## Apriori Model

```python
book_frequent_itemsets_model1 = apriori(book_data, min_support=0.1, use_colnames=True)
book_frequent_itemsets_model1
```

### In above data set support value minimum than 0.1 not consider.

```python
book_rules_model1 = association_rules(book_frequent_itemsets_model1, metric='lift', min_threshold=0.2)
book_rules_model1
```

### If the leverage value is 0 then it indicates independents. Also, high conviction value means that the consequents is highly depending on the antecedents.

```python
book_rules_model1.sort_values('lift', ascending=False).head(20)
```

```python
book_rules_model1[book_rules_model1.lift>1]
```

### We know that if the lift ratio high i.e. greater than 1 then it is the best rule. So that,
1. If someone purchased only Youth books then he or she will also purchase Child book also.
2. If someone purchased only Cook books then he or she will also purchase Geogbook and Art book also.
### Hence, we can conlude this type of rules.
