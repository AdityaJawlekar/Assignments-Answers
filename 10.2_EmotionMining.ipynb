# Q. Emotion Mining on reviews from ecommerce website Amazon  

## Importing Libraries

```python
import numpy as np
import pandas as pd 
import spacy
from matplotlib import pyplot as plt
import string
from nltk import tokenize
```

## Loading Data Sets

```python
review = pd.read_csv('apple.txt',error_bad_lines=False)
review = [x.strip() for x in review.x]
review = [x for x in review if x]
afinn = pd.read_csv('Afinn.csv',sep=',', encoding='latin-1')
```

```python
sentences = tokenize.sent_tokenize(" ".join(review))
```

```python
review_df = pd.DataFrame(sentences, columns=['sentence'])
review_df
```

```python
affinity_scores = afinn.set_index('word')['value'].to_dict()
nlp = spacy.load('en_core_web_sm')
sentiment_lexicon = affinity_scores

def calculate_sentiment(text: str = None):
    sent_score = 0
    if text:
        sentence = nlp(text)
        for word in sentence:
            sent_score += sentiment_lexicon.get(word.lemma_, 0)
    return sent_score
```

```python
review_df['sentiment_value'] = review_df['sentence'].apply(calculate_sentiment)
review_df.sort_values(by='sentiment_value')
review_df
```
