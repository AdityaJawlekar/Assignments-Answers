# Answer of Q11

```python
from scipy import stats
```

```python
avg_wt = stats.norm.interval(0.97, loc = 200, scale = 30)
print("Person weight at 94% confidence interval:", np.round(avg_wt,2))
```

```python
avg_wt = stats.norm.interval(0.99, loc = 200, scale = 30)
print("Person weight at 99% confidence interval:", np.round(avg_wt,2))
```

```python
avg_wt = stats.norm.interval(0.98, loc = 200, scale = 30)
print("Person weight at 96% confidence interval:", np.round(avg_wt,2))
```

# Answer of Q20

```python
cars_df = pd.read_csv("Cars.csv")
cars_df
```

```python
print("Probability of MPG more than 38 is:",round(1-stats.norm.cdf(38,loc = cars_df.MPG.mean(), scale = cars_df.MPG.std()),4))
```

```python
print("Probability of MPG less than 40 is:",round(stats.norm.cdf(40,loc = cars_df.MPG.mean(), scale = cars_df.MPG.std()),4))
```

```python
print("Probability of MPG less than 20 and more than 50 is:",round(stats.norm.cdf(50,loc = cars_df.MPG.mean(), scale = cars_df.MPG.std())-stats.norm.cdf(20,loc = cars_df.MPG.mean(), scale = cars_df.MPG.std()),4))
```

# Answer of Q21(b)

```python
X = pd.DataFrame(columns = ["wtat_df.Waist","wtat_df.AT"]) 
X["wtat_df.Waist"] = pd.Series(wtat_df.Waist) 
X["wtat_df.AT"] = pd.Series(wtat_df.AT) 
sns.distplot(X['wtat_df.AT'],label = 'AT'); 
sns.distplot(X['wtat_df.Waist'],label = 'Waist')
plt.xlabel('AT & Waist');
plt.ylabel('Density');
plt.legend();
```
