# Q2

```python
## More employees at the processing center are older than 44 than between 38 and 44
prob_older_than_44 = round(1-(stats.norm.cdf(44,loc=38,scale=6)),4)
prob_older_than_44
prob_between_38_and_44 = round((stats.norm.cdf(44,loc=38,scale=6)) - (stats.norm.cdf(38,loc=38,scale=6)),4)
prob_between_38_and_44
print("More employees at the processing center are older than 44 is:", prob_older_than_44)
print("More employees at the processing center are between 38 and 44 is:", prob_between_38_and_44)
```

```python
## training program for employees under the age of 30 at the center would be expected to attract about 36 employees
prob_under_30 = round(stats.norm.cdf(30,loc=38,scale=6),4)
print("Employees under the age of 30 is:", prob_under_30)
```
